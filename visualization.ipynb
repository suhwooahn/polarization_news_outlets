### 1. 언론사 및 날짜 선택

```python
dt1 = dt[dt['article_outlet'].str.contains('조선일보')]

#resample time-series data with pandas

grouped1 = dt1["new_labels"].groupby([dt1["president_year_2"], dt1["category"]])
new_dt1 = grouped1.mean().unstack("category")
reset1 = new_dt1.reset_index()
reset1 = reset1.sort_values(["president_year_2"], ascending=[True])
```

### 1.1 그래프 그리기

```python
dt = pd.read_csv('C:/Users/Ahn Suhwoo/Desktop/Political Com/Partisan media/python_text/visual_data.csv', encoding='utf-8')
dt.head()
```

```python
dt1 = dt[dt['category'].str.contains('신규')]

#resample time-series data with pandas

grouped1 = dt1["distance1"].groupby([dt1["quarter"], dt1["category"]])
new_dt1 = grouped1.mean().unstack("category")
reset1 = new_dt1.reset_index()
reset1 = reset1.sort_values(["quarter"], ascending=[True])
reset1
```

```python
dt2 = dt[dt['category'].str.contains('기성')]

#resample time-series data with pandas

grouped2 = dt2["distance1"].groupby([dt2["quarter"], dt2["category"]])
new_dt2 = grouped2.mean().unstack("category")
reset2 = new_dt2.reset_index()
reset2 = reset2.sort_values(["quarter"], ascending=[True])
reset2
```

```python
x1 = list(reset1['quarter'])
y1 = list(reset1['신규'])

x2 = list(reset2['quarter'])
y2 = list(reset2['기성'])

n1 = ['12:1', '12:2', '12:3', '12:4', '13:1', '13:2', '13:3', '13:4', '14:1', '14:2', '14:3', '14:4', '15:1', '15:2', '15:3', '15:4',\
      '16:1', '16:2', '16:3', '16:4', '17:1', '17:2', '17:3', '18:1', '18:2', '18:3', '18:4']
n2 = ['01:01', '', '', '', '02:01', '', '', '', '03:01', '', '', '', '04:01', '', '', '', '05:01', '', '', '', '06:01', '', '', '',\
      '07:01', '', '', '', '08:01', '', '', '', '09:01', '', '', '', '10:01', '', '', '', '11:01', '', '', '', '12:01', '', '', '',\
      '13:01', '', '', '', '14:01', '', '', '', '15:01', '', '', '', '16:01', '', '', '', '17:01', '', '', '18:01', '', '', '']
```

```python
fig = plt.figure(figsize=(15, 10))
#plt.title("Partisan Score of the Media Outlets", fontsize=25, y=1.05)

#plt.axvspan(1, 28, facecolor='yellow', alpha=0.5)
#plt.axvspan(65, 71, facecolor='yellow', alpha=0.5)

#1
plt.plot(x1, y1, c='red', label="신규 언론사", lw=2, ls="-",
         marker=".", ms=15, mec="lightcoral", mfc="lightcoral")
#2
plt.plot(x2, y2, c='blue', label="기성 언론사", lw=2, ls="-",
         marker=".", ms=15, mec="darkcyan", mfc="darkcyan")
#3
#plt.plot(x1, y1, c='green', label="Established media outlets", lw=2, ls="-",
#         marker=".", ms=15, mec="mediumseagreen", mfc="mediumseagreen")

#plt.xticks(range(45,72,1), [str(i).zfill(2) for i in range(45, 72,1)], fontsize=14, rotation=45)
plt.xticks(range(45,72,1), n1, fontsize=14, rotation=45)
plt.yticks([0, 5, 10, 15], fontsize=14)

plt.grid(True)

plt.legend(loc=2, prop={'size': 20})

plt.xlabel("연도: 분기", fontsize=20, labelpad=30)
plt.ylabel("이념적 극단성 점수", fontsize=20, labelpad=30)

plt.savefig("중앙일보와 JTBC 논조 차이")
plt.show()
```
